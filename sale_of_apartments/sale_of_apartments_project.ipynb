#### Пропуски по столбцу населенного пункта `locality_name`
Отсутствие населенного пункта в объявлении непреемлемо, это может ввести потенциального покупателя в заблуждение и исказить наш анализ. Выясним какую долю они составляют в общей базе.

```python
#определяем долю пропущенных значений
null_locality = len(df[df['locality_name'].isnull()]) / len(df['locality_name'])
print('Доля незаполненых населенных пунктов:',f'{null_locality: .2%}')
```

Доля таких объявлений существенно мала, поэтому удалим их из таблицы, где через логическое индексирование отсечем незаполненые строки. Можно также воспользоваться методом .drop(), но этот метод проще.

```python
#обновляем таблицу и сбрасываем индексы с помощью `reset_index(drop=True)`
df = df[~df['locality_name'].isnull()].reset_index(drop=True)
#проверка
null_locality = len(df[df['locality_name'].isnull()]) / len(df['locality_name'])
print('Доля незаполненых населенных пунктов после обработки:',f'{null_locality: .2%}')
```

☝ Так как столбец `locality_name` может быть ключевой переменной при дальнейших расчетах (н-р медианы) в заполнении пропусков остальных столбцов, проверим насколько корректно заполнены названия населенных пунктов.

```python
print(list(df['locality_name'].sort_values().unique()))
```

Мы видим некоторые проблемы, самое бросающиеся в глаза это "е-ё", а также разные написания "городской посёлок - посёлок городского типа", это будет наводит путаницу и дублировать одну и ту же локацию.
<br>Для исправления данной ситуции, очистим в названиях населенных пунктов их типы: "село", "поселок" и т.д. 
<br>***Цель данной процедуры*** - получить уникальные значения населенных пунктов т.к. как отметили выше возникает путаница, помимо очевидных проблемых как "е-ё", также имеются более серьезные нарушения как разные типы одной локации, н-р "деревня Рабитицы - поселок Рабитицы", которая по факту является деревней.
<br>Мы не будем прибегать к лемматизации и отбирать слова, здесь можно обойтись проще, так как все названия населенных пунктов указаны с заглавной буквы, мы удалим все слова, которые написаны с маленькой буквы, а это как раз будет типами.

```python
#создадим функцию, которая будет удалять слова не имеющие букв с верхним регистром
def del_islower(s):
    return ' '.join(w for w in s.split(' ') if not w.islower())

#применим функцию к столбцу с населенным пунктами одной строкой с помощью .apply()
df['locality_name'] = df['locality_name'].apply(del_islower)

#проверим как изменилось написание
print(list(df['locality_name'].sort_values().unique()))
```

Как мы видим, все населенные пункты приведены к нужному виду, теперь их можно использовать в анализе.

#### Заполнение пропусков по столбцу высоты потолков `ceiling_height`
Высота потолка, может повлиять на решение покупателя приобретать квартиру или нет, а также может оказать влияние на цену квартиры.
<br> Заполним пропуски медианными значениями в разрезе городов т.к. с большей вероятность в каждом городе дома с типовыми застройками и одинаковыми высотами потолков.

```python
#заполним пропуски медианными значениями с помощью методов fillna() и transform()
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('locality_name')['ceiling_height'].transform('median'))
#также учтем момент, что не по всем городам вообще указывали высоту, для таких населенных пунктов укажем высоту 2.6 (средняя по стране) 
df['ceiling_height'] = df['ceiling_height'].fillna(2.6)
#проверка
print('Количество незаполненных значений:', df['ceiling_height'].isna().sum())
```

#### Заполнение пропусков по жилой площади `living_area` и площади кухни `kitchen_area`
В представленных данных у нас имеется общая площадь квартиры, но по многим объявлениям не заполнены жилая площадь и/или площадь кухни.
Будем выполнять заполнение поэтапно.

```python
#Для студий `is_studio` укажем площадь кухни `kitchen_area` = 0 т.к. кухня как правило совмещена с жилой комнатой
df.loc[df['is_studio']==True,['kitchen_area']] = 0
#Также проверим, если студия, то заполним жилую площадь `living_area` как = `общая площадь`-`30% от общей (прочая площадь)`
#под прочей понимаем площадь санузла/коридора/балкона и пр.
df['living_area'] = df['living_area'].fillna(df[df['is_studio']==True]['total_area'] * 0.7)
```

```python
#площадь кухни посчитаем как = `общая площадь`-`жилая площадь` методом .sub()
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'].sub(df['living_area'], fill_value = 0))
```

```python
#жилую площадь `living_area` посчитаем как =`общая площадь`-`30% от общей (прочая площадь)`- `площадь кухни`
#для вычитания используем метод .sub() с параметром fill_value = 0, в случае если площадь кухни нулевая
df['living_area'] = df['living_area'].fillna(df['total_area'].sub(df['kitchen_area'], fill_value = 0) * 0.7)
```

#### Заполнение пропусков по столбцу количества этажей в доме `floors_total`
Если зайти на какой-либо сайт застройщика, где можно посмотреть квартиры с планом/ценой/расположением, то с большей вероятностью вы обнаружите, что первый и последние этажи, особенно высоток часто продаются в последнюю очередь. Поэтому покупателю важно понимать, на каком этаже будет находиться его квартира относительно этажности всего дома.
<br> 
<br> Для начала проверим, какой процент объявлений занимает в базе, где не заполнена этажность дома.

```python
#количество пропущенных значений
null_floors = len(df[df['floors_total'].isnull()]) / len(df['floors_total'])
print('Доля незаполненых населенных пунктов после обработки:',f'{null_floors: .2%}')
```

Учитывая, что доля объявлений с незаполненной этажностью дома составляет 0.36%, мы можем их удалить.

Наличие балкона является безусловным плюсом, но в наших данных пустые значения могут означать, что поле не является обязательным к заполнению, поэтому предположим, что у данных квартир балконы отсутствуют

Заполнение данного столбца необходим, что в дальнешем по нему корректно определялся его булевой тип, без этого будет тип `object` и если нам в анализе потребуется данных столбец, будет корректно обращаться к нему в одном формате

### Изменение типов данных
Использование корректных типов данных предоставляет дополнительную возможность экономить определенное количество памяти, что в целом скажется на производительности проведения анализа и более удобному интерпретированию результатов.

```python
#создадим функцию для определения занимамемой памяти
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else: # исходим из предположения о том, что если это не DataFrame, то это Series
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024 ** 2 # преобразуем байты в мегабайты
    return "{:03.2f}".format(usage_mb)
#проверим, сколько памяти у нас занимает таблица до преобразований и в конце отобразим результат после
#создадим переменную mem_before, чтобы в конце сравнить, сколько мы сэкономили места
mem_before = mem_usage(df)
print('Таблица до изменений типов данных занимает в памяти:',mem_before)
```

#### Сконвертируем столбцы с `float` в `int`
Некоторые столбцы нет смысла использовать с плавающей точкой т.к. по своему назначению они должны отражать целые числа.
<br>Так как большая часть столбцов должны быть целочисленными, то возьмем все `float`, кроме тех что должны остаться с данным типом и сконвертируем их в `int`

```python
#столбцы float, которые останутся с этим форматом p.s 'last_price' в том числе т.к. мы сконвертировали в млн
c_exception = ['last_price','total_area','ceiling_height','living_area','kitchen_area']
#перед конвертацией заполним пропуски в 0 и применим конвертацию в целое с параметром downcast, который понизит до минимального битность
df[df.select_dtypes(include=['float']).columns.difference(c_exception)] = df[df.select_dtypes(include=['float']).columns.difference(c_exception)].apply(pd.to_numeric,downcast='integer')
```

#### Сконвертируем столбец даты и времени `first_day_exposition`
Для извлечения составляющий из даты, например дня, необходимо сконвертировать столбец в формат `datetime` и в дальнейшем мы сможем извлекать день с помощью метода `.dt.day`

```python
#с помощью .to_datetime(, format='') сконвертируем столбец в формат `datetime`
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y.%m.%d')
```

#### Оптимизация подтипов данных
При загрузки таблицы Python по умолчанию выставляет типы данных в 64 бита т.е. на максимум, но по факту в столбце могут быть числа (если числовой тип) с максимальным значением 10, таким образом мы можем оптимизировать таблицу, снизим битность подтипа с 64 до 8. 
<br><br>Тоже касается и текстовых данных, столбцы по ним имеют тип `object` и при большом объеме данный тип данных может сильно отразиться на объеме занимаемой памяти, в случае если в столбце уникальных значений менее 50% от общего объема, можно сконвертировать в тип `category`, где уникальные значения будут храниться в отдельном словаре Pandas и устанавливать соответствия целочисленных и исходных значений.

```python
#оптимизация по типу данных 'int'
df[df.select_dtypes(include=['int']).columns] = df.select_dtypes(include=['int']).fillna(0).apply(pd.to_numeric,downcast='integer')

#оптимизация по типу данных 'float'
df[df.select_dtypes(include=['float']).columns] = df.select_dtypes(include=['float']).apply(pd.to_numeric,downcast='float')

#оптимизация по типу данных 'object'
df[df.select_dtypes(include=['object']).columns] = df.select_dtypes(include=['object']).astype('category')
```

```python
#проверка
mem_after = mem_usage(df)
mem_effect = float(mem_before)/float(mem_after)
print('Таблица после изменений типов данных занимает в памяти:', mem_after)
print('Эффект от оптимизации сотавил:', f'{mem_effect: .2%}')
```

Как правило, анализ недвижимости мог бы рассматриваться не только по всем объектам, но и по уровню недвижимости (бизнес, комфорт, эконом). В рамках данного проекта было бы правильно рассматривать наиболее доступную недвижимость т.к. предложений по ней больше и мы сможем более детально проанализировать взаимосвязи между параметрами, поэтому расчеты по отсечению выбросов будем выполнять в основном по усам диаграммы рассеивания.
<br>C помощью метода `.query()` отфильтруем данные по усам, которые определим с помощью функции `out_r()`.

```python
#функция для определения нижнего и верхнего уса
def out_r(data):
    Q1 = data.quantile(q=0.25)
    Q3 = data.quantile(q=0.75)
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
#создадим новую таблицу 'df_n', где будут храниться данные очищенные от выбросов
df_n = df.query(
    '''
    @out_r(last_price)[0] <= last_price <= @out_r(last_price)[1] & \
    @out_r(total_area)[0] <= total_area <= @out_r(total_area)[1] & \
    @out_r(rooms)[0] <= rooms <= @out_r(rooms)[1] & \
    @out_r(ceiling_height)[0] <= ceiling_height <= @out_r(ceiling_height)[1] & \
    @out_r(days_exposition)[0] <= days_exposition <= @out_r(days_exposition)[1]
    '''
).reset_index(drop=True)
#уровень очистки
print('Таблица очищена от выбросов на',f'{1-len(df_n)/len(df):.2%}')
```

### Исследование факторов, влияющих на стоимость квартиры

Рассмотрим влияние на цену таких показателей как: площадь, число комнат, удаленность от центра, этаж и дата размещения.
<br>Для нахождения взаимосвязи мы используем метод `.corr()` для определения коэффициента корреляции Пирсона, а также построим диаграмму рассеивания `scatter` для визуального обнаружения связи. 
<br>Для удобства создадим функцию `create_multichart()`, которая автоматизирует формирование графиков (до двух) и рассчитает коэф.корреляции с отображением на графике.

```python
#функция создания одинарного/двойного графика scatter или line, !!! наименование столбцов в двух таблицах должны быть одинаковыми
#p.s. функцию можно еще оптимизировать т.к. есть повторения в коде, но из-за ограниченного времени оставлю пока так, в след проекты оптимизирую
def create_multichart(df_in, col_x,col_y,name_x,name_y,type_chart, df2_in = None):
    #если мы делаем двойной график, то строим параметры для второго, потом для первого
    if df2_in is not None:
        fig, axes = plt.subplots(1, 2)
        ax_df1 = axes[0]
        ax_df2 = axes[1]
        n=15 #ширина графика
        avg_df2 = df2_in.pivot_table(index=col_x, values=col_y, aggfunc='mean').reset_index()
        if type_chart=='sc': avg_df2.plot(ax=ax_df2, x=col_x, y=col_y, kind='scatter', figsize=(n, 5), grid=True, alpha=0.5)
        if type_chart=='line': avg_df2.plot(ax=ax_df2, x=col_x, y=col_y, kind='line', figsize=(n, 5), grid=True, legend=False,style='o-')
        corr_v2 = avg_df2[col_y].corr(avg_df2[col_x].astype('category').cat.codes)
        ax_df2.set_title('Коэф.корр: {:.2}'.format(corr_v2), fontsize=10); ax_df2.set_xlabel(name_x, fontsize=10); ax_df2.set_ylabel(name_y, fontsize=10)
    else:
        #если у нас одинарный график, то параметры для отображения 1 графика
        fig, axes = plt.subplots()
        ax_df1 = axes
        n = 7 #ширина графика
    #продолжаем построение
    avg_df1 = df_in.pivot_table(index=col_x, values=col_y, aggfunc='mean').reset_index()
    #переменная графика
    if type_chart=='sc': avg_df1.plot(ax=ax_df1, x=col_x, y=col_y, kind='scatter', figsize=(n, 5), grid=True, alpha=0.5)
    if type_chart=='line': avg_df1.plot(ax=ax_df1, x=col_x, y=col_y, kind='line', figsize=(n, 5), grid=True, legend=False,style='o-', sharex=True)
    #добавим на график текст с коэф.корреляции c группировкой по средней + .astype('category').cat.codes если будут строковые значения
    corr_v1 = avg_df1[col_y].corr(avg_df1[col_x].astype('category').cat.codes)
    #заголовок и названия осей
    plt.suptitle("Зависимость цены от параметра: " + name_x, fontsize=14)
    ax_df1.set_title('Коэф.корр: {:.2}'.format(corr_v1), fontsize=10); ax_df1.set_xlabel(name_x, fontsize=10); ax_df1.set_ylabel(name_y, fontsize=10)
    #подписи на график
    if df2_in is not None:
        x_min1 = 0 if df_in[col_x].dtype == 'object' else avg_df1[col_x].min()
        ax_df1.text(x_min1, avg_df1[col_y].max(), ' Регион',color='blue',fontsize=12)
        x_min2 = 0 if df_in[col_x].dtype == 'object' else avg_df2[col_x].min()
        ax_df2.text(x_min2, avg_df2[col_y].max(), ' Центр Питера',color='blue',fontsize=12)
    #отображаем график
    plt.show()
```

```python
#строк всего 14, поэтому удалим их
df_spb = df_spb[~pd.isnull(df_spb['city_centers_nearest'])]
#сконвертируем расстояние из метров в километры и переведем в целые
df_spb['city_centers_nearest_km'] = (df_spb['city_centers_nearest']/1000).astype('int')
```

```python
#сформируем таблицу со средними значениями по каждому километру в параметрах цены и площади
#p.s график можно было построить сразу черед сводную, но захотел добавить заливку, поэтому создал промежуточное звено df_ax, не ругайте :)
df_ax = df_spb.pivot_table(index='city_centers_nearest_km',values=['last_price'],aggfunc='mean').reset_index()
ax = df_ax.plot(x='city_centers_nearest_km', y='last_price', kind='line', figsize=(15, 5), grid=True
    ,stacked = True, xticks=range(0,21),yticks=range(0,8), xlim = (0,20), legend=False,style='.-')
#добавим заголовок и подписи осей
ax.set_title('Изменение цены в зависимости от удаленности от центра', fontsize=12)
ax.set_xlabel('Удаленность от центра', fontsize=10)
ax.set_ylabel('Стоимость в млн.', fontsize=10)
#добавим подпись центральной зоны
ax.text(2, 3.2, 'Центральная зона',color='green',fontsize=12)
#сделаем заливку центральной зоны
plt.fill_between(range(8), df_ax[df_ax['city_centers_nearest_km']<=7]['last_price'], color='green', alpha=0.2)
plt.fill_between(range(7,21), df_ax[(df_ax['city_centers_nearest_km']>=7) & (df_ax['city_centers_nearest_km']<=20)]['last_price'], color='yellow', alpha=0.2)
#отображаем график
plt.show()
```

Исходя из данного графика, можно предположить, что изменение стоимости квартиры наиболее чувствуется, начиная с 7 км от центра, далее цена снижается и остается более-менее стабильной. Таким образом, сделаем вывод, что центральной зоной (выделил заливкой) в Питере является расстояние до 7 км от центра.

### Сравнение выводов по квартирам в центре и общих выводов по всему городу

В начале исследования мы создавали функцию, которая строила гистограммы по параметру из таблицы. Для текущего исследования модернизируем немного нашу функцию и добавим наслоение двух массивов. Наша задача посмотреть как распределяются данные по корзинам.

```python
def create_chart(df, df2, col, col_name, b, rng):
    #предопределяем график, что будет возможность его дополнить сверху
    fig, ax1 = plt.subplots(figsize=(14,4))
    ax2=ax1.twinx()
    #добавляем первую гистограмму
    df.plot(ax=ax1, y=[col], kind='hist', bins=b, range=rng, grid=True,legend=True, alpha = 0.3, label = ['Регион'], color="r")
    #добавляем вторую гистограмму
    df2.plot(ax=ax2, y=[col], kind='hist',bins=b, range=rng, legend=True, alpha = 0.3, color="b", label = ['Центр'])
    #добавим заголовок и подписи осей
    ax1.axes.set_title(col_name, fontsize=14)
    ax1.set_xlabel(col_name, fontsize=10)
    ax1.set_ylabel("Итого количество объявлений", fontsize=10)
    ax2.set_ylabel("Количество объявлений в центре", fontsize=10)
    ax1.legend(loc=3, bbox_to_anchor= (1.07, 0.9), ncol=1,
            borderaxespad=0, frameon=False)
    ax2.legend(loc=2, bbox_to_anchor= (1.07, 0.9), ncol=1,
            borderaxespad=0, frameon=False)
    plt.show()
```
